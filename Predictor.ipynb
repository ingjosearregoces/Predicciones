IMPORTAR LIBRERIAS

```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,precision_score,recall_score,confusion_matrix
from sklearn.metrics import roc_curve, auc
from sklearn import metrics 
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn import metrics
from sklearn.metrics import matthews_corrcoef
from sklearn.preprocessing import Normalizer
```

DATOS

```
datosTitanic = pd.read_csv('/content/train.csv')
datosTitanic.head()
```

ELIMINACION COLUMNAS INNECESARIAS

```
datosTitanic = datosTitanic.drop(columns = ['Cabin','PassengerId','Name','Ticket'],axis=1)
datosTitanic
```

REEMPLAZO DE VALORES SIMPLIFICADOS

```
datosTitanic.replace({'Embarked':{'S':0,'C':1,'Q':2}},inplace=True)
datosTitanic.replace({'Sex':{'male':0,'female':1}},inplace=True)
datosTitanic
```

AMPLIACION DE VALORES

```
datosTitanic['Embarked'].fillna(datosTitanic['Embarked'].mode()[0], inplace=True)
datosTitanic['Pclass'].fillna(datosTitanic['Pclass'].mode()[0], inplace=True)
datosTitanic['Sex'].fillna(datosTitanic['Sex'].mode()[0], inplace=True)
datosTitanic['Age'].fillna(datosTitanic['Age'].mode()[0], inplace=True)
datosTitanic['SibSp'].fillna(datosTitanic['SibSp'].mode()[0], inplace=True)
datosTitanic['Parch'].fillna(datosTitanic['Parch'].mean(), inplace=True)
datosTitanic['Fare'].fillna(datosTitanic['Fare'].mean(), inplace=True)
datosTitanic
```

NORMALIZACION

```
norma = Normalizer().fit(datosTitanic)
normalizado=norma.transform(datosTitanic)
normalizado
```

DATA FRAME

```
dataframe=pd.DataFrame(normalizado ,columns=['Sobrevivio','ClasePasajero','Sexo','Edad','Hermanos','Padres','Costo','Embarco'])
dataframe
```

```
Y=datosTitanic['Survived']
X=dataframe.drop(columns=['Sobrevivio'], axis=1)
X
```

```
Y
```

```
Xtrain, Xtest, Ytrain, Ytest = train_test_split(X, Y, random_state=4)
```

KNN

```
ClasificadorKNN = KNeighborsClassifier()
ClasificadorKNN.fit(Xtrain,Ytrain)
Ypred = ClasificadorKNN.predict(Xtest)
PresicionKNN = round(accuracy_score(Ytest, Ypred), 6)
print("Presicion:",PresicionKNN*100,"%")
print("MCC:",matthews_corrcoef(Ytest, Ypred))
```

REGRESION

```
regresion = LogisticRegression()
regresion.fit(Xtrain,Ytrain)
Ypred1 = regresion.predict(Xtrain)
Presicion = round(metrics.accuracy_score(Ytrain, Ypred1), 5)
PorcentajePresicion = 100 * Presicion 
print("PResicion:", PorcentajePresicion,"%")
print("MCC:",matthews_corrcoef(Ytrain, Ypred1))
```

```
regresion.fit(Xtest,Ytest)
Ypred2 = regresion.predict(Xtest)
Presicion1 = round(metrics.accuracy_score(Ytest, Ypred2), 6)
PorcentajePresicion1 = 100 * Presicion1 
print("Presicion:", PorcentajePresicion1,"%")
print("MCC:",matthews_corrcoef(Ytest, Ypred2))
```

SVM

```
fichero = SVC(kernel='linear')
fichero.fit(Xtrain, Ytrain)
Ypred3 = fichero.predict(Xtest)
Presicion2 = round(metrics.accuracy_score(Ytest, Ypred3),6)
PorcentajePresicion2 = 100 * Presicion2 
print("presicion:", PorcentajePresicion2,"%")
print("MCC:",matthews_corrcoef(Ytest, Ypred3))
```
